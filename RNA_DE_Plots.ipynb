```python
toptable = pd.read_csv('Prim-Meta_TopTable-True.csv')
toptable = toptable.set_index('X', drop=True)
toptable = toptable.loc[toptable.index.isin(cod_genes)]
toptable.head()
```

```python
toptable = toptable.loc[toptable['AveExpr'] > 10]
```

```python
toptable.loc['HLA-E']
```

```python
nonsig = toptable.loc[(toptable['adj.P.Val'] > 0.05) | (np.abs(toptable['logFC']) < 1)]
sig = toptable.loc[(toptable['adj.P.Val'] < 0.05) & (np.abs(toptable['logFC']) > 1)]

high_met = sig.loc[sig['logFC'] < 0]
high_pri = sig.loc[sig['logFC'] > 0]
```

```python
m = min(sig['logFC'])
sig.loc[sig['logFC'] == m]
```

```python
p = bokeh.plotting.figure(
    width=1000,
    height=1000,
    title='Volcano Plot of Primary vs. Metastasis',
    x_axis_label = 'log2(FC)',
    y_axis_label='-log10( Adjusted P Value)',


)

p.circle(
    x = nonsig['logFC'],
    y=-np.log10(nonsig['adj.P.Val']),
    color = 'gray'
)

p.circle(
    x = high_pri['logFC'],
    y=-np.log10(high_pri['adj.P.Val']),
    color = 'blue',
    legend_label = 'Upregulated in Primary Tumors'
)

p.circle(
    x = high_met['logFC'],
    y=-np.log10(high_met['adj.P.Val']),
    color = 'red',
    legend_label = 'Upregulated in Metastatic Tumors'
)

p.line(
    x=[-1.5e5, 1.5e5],
    y=[-np.log10(0.05), -np.log10(0.05)],
    line_dash = 'dashed'
)
p.line(
    x=[-1,-1],
    y=[0,12],
    line_dash='dashed'
)

p.line(
    x=[1,1],
    y=[0,12],
    line_dash='dashed'
)


bokeh.io.show(p)
```

```python
def gene_to_prism(df, gene, folder = 'Prism_CSV'):
    norm = df.loc[df['Type'] == 'Normal'].reset_index(drop=True)
    met = df.loc[df['Type'] == 'Metastatic'].reset_index(drop=True)
    pri = df.loc[df['Type'] == 'Primary'].reset_index(drop=True)
    
    new_df = pd.DataFrame({'index':range(1,100)})
    new_df['Normal'] = norm[gene]
    new_df['Metastasis'] = met[gene]
    new_df['Primary'] = pri[gene]
    
    new_df.to_csv(f'{folder}/{gene}.csv')
```
